### Premissas

```python
# Dados da GOL (GLO), Azul (AZU) e LATAM (TAM)
df = df[(df['ICAO Empresa Aérea']=='GLO') |
        (df['ICAO Empresa Aérea']=='AZU') |
        (df['ICAO Empresa Aérea']=='TAM')]
```

```python
# Remove voos internacionais
df = df[(df['Código Tipo Linha']!='I') &
        (df['Código Tipo Linha']!='G')]
```

```python
# Remove coluna "Código Justificativa"
df = df.drop('Código Justificativa', axis=1)
```

```python
# Remove valor "Cancelado" da coluna "Situação Voo"
# Remove coluna "Situação Voo"
df = df[df['Situação Voo'] != 'CANCELADO']
df = df.drop('Situação Voo', axis=1)
```

```python
# Remove valores duplicados e valores nulos
df = df.drop_duplicates()
df = df.dropna()
```

```python
# Transformar datas para timestamp
df['Partida Real timestamp'] = [
    datetime.strptime(i, '%d/%m/%Y %H:%M').timestamp() for i in df['Partida Real']
]

df['Partida Prevista timestamp'] = [
    datetime.strptime(i, '%d/%m/%Y %H:%M').timestamp() for i in df['Partida Prevista']
]

df['Diferença de Tempo'] = df['Partida Real timestamp'] - df['Partida Prevista timestamp']
```

```python
# Análise de dados do tempo
print('Media:', round(df['Diferença de Tempo'].mean(), 2))
print('Diferença de Tempo Max:', df['Diferença de Tempo'].max())
print('Diferença de Tempo Min:', df['Diferença de Tempo'].min())

print('\nBoxplot')
df['Diferença de Tempo'].plot.box()
```

### Pela análise do boxplot, alguns dados são outliers

```python
# Remove outliers
df = df[df['Diferença de Tempo'] >= -1000000]
df = df[df['Diferença de Tempo'] <= 5000000]
```

```python
# Remove colunas que não serão utilizadas
df = df.drop(['Chegada Prevista', 'Chegada Real', 'Partida Real timestamp', 'Partida Prevista timestamp'], axis=1)
```

```python
# 0 - Voo atrasado
# 1 - Voo pontual
df['Classificação'] = [1 if i <= 0 else 0 for i in df['Diferença de Tempo']]
```

# Etapa 2 – Implementação de um modelo de classificação para prever atraso do voo

```python
df_model = df_situacao_voo[
    ['ICAO Empresa Aérea',
     #'Número Voo',
     'Código Autorização (DI)',
     #'Código Tipo Linha',
     #'ICAO Aeródromo Origem',
     #'ICAO Aeródromoo Destino',
     'Partida Prevista Mes',
     'Partida Prevista Hora',
     'Classificação'
    ]
]
```

```python
df_model = df_model[df_model['Código Autorização (DI)']!='B']
df_model['Código Autorização (DI)'] = df_model['Código Autorização (DI)'].astype(int)
```

```python
df_model = pd.get_dummies(df_model)
```

```python
df_model
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import plot_confusion_matrix, plot_roc_curve
from sklearn.metrics import confusion_matrix

from matplotlib import pyplot
from numpy import mean
from numpy import std

# define dataset
X = df_model.drop('Classificação', axis=1)
y = df_model['Classificação']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 42)

# define the model
model = RandomForestClassifier(n_estimators = 10, random_state = 42)
```

```python
# fit the model
model.fit(X_train, y_train)
```

```python
# get importance
importance = model.feature_importances_
```

```python
# summarize feature importance
for i,v in enumerate(importance):
    print('Feature: %0d, Score: %.5f' % (i,v))
```

```python
# plot feature importance
pyplot.bar([x for x in range(len(importance))], importance)
pyplot.show()
```

```python
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=42)
n_scores = cross_val_score(model, X_train, y_train, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
# report performance
print('Accuracy: %.3f (%.3f)' % (mean(n_scores), std(n_scores)))
```

```python
y_pred = model.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
plot_confusion_matrix(model, X_test, y_test)  
```

```python
plot_roc_curve(model, X_test, y_test)
```

# Compra de novas aeronaves

### Variáveis de Decisão

- X1 -> Quantidade de aviões Ac
- X2 -> Quantidade de aviões Am
- X3 -> Quantidade de aviões Al

### Função Objetivo

- L = 190 * X1 + 220 * X2 + 310 * X3

### Restrições

- 2.1 * X1 + 4 * X2 + 5.2 * X3 <= 120
- X1 + X2 + X3 <= 25
- X1 + 4/3 * X2 + 5/3 * X3 <= 36
- X3 + X2 >= X1
- X1 + X2 >= 7 * X3
- X1 >= 0
- X2 >= 0
- X3 >= 0

```python
L = 0

for X1 in list(range(26)):
    for X2 in list(range(26)):
        for X3 in list(range(26)):
            if ((2.1 * X1 + 4 * X2 + 5.2 * X3 <= 120) and
                (X1 + X2 + X3 <= 25) and
                (X1 + 4/3 * X2 + 5/3 * X3 <= 36) and
                (X3 + X2 >= X1) and
                (X1 + X2 >= 7 * X3) and
                (X1 >= 0) and
                (X2 >= 0) and
                (X3 >= 0)):
                    L1 = 190 * X1 + 220 * X2 + 310 * X3
                    if L1 > L:
                        L = L1
                        x1 = X1
                        x2 = X2
                        x3 = X3

print('Ac:', x1)
print('Am:', x2)
print('Al:', x3)
print('Lucro:', L)
```
